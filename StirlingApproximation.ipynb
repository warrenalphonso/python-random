```python
import math
import scipy as sp
import matplotlib.pyplot as plt
```

```python
n = 100000
xs = range(1, n, n // 20)
ys = [math.log(sp.math.factorial(x)) for x in xs]
stirling = [x * math.log(x) for x in xs]
```

```python
plt.plot(xs, ys, linestyle="--", marker="o", color="b", label="$\ln(n!)$")
plt.plot(xs, stirling, linestyle="--", marker="o", color="r", label="$n\ln(n)$")
plt.legend()
plt.show()
```

Turns out `np.log` complains for decently large numbers but `math.log` doesn't break a sweat.
